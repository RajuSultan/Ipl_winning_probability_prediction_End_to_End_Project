```python
import numpy as np
import pandas as pd
```

```python
match=pd.read_csv("matches.csv")
delivery=pd.read_csv("deliveries.csv")
```

```python
match.head()
delivery.head()
```

```python
total_score_df=delivery.groupby(["match_id","inning"]).sum()["total_runs"].reset_index()
total_score_df['total_runs'] = total_score_df['total_runs'] + 1
total_score_df
```

```python
total_score_df = total_score_df[total_score_df["inning"]== 2]
total_score_df
```

```python
match_df=match.merge(total_score_df[['match_id', 'total_runs']], left_on='id', right_on='match_id')
match_df
```

```python
match_df['team1'].unique()
```

```python
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]
```

```python
match_df['team1']=match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2']=match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df=match_df[match_df['team1'].isin(teams)]
match_df=match_df[match_df['team2'].isin(teams)]
```

```python
match_df = match_df[match_df['dl_applied']==0]
match_df
```

```python
match_df = match_df[['match_id','city', 'winner', "total_runs"]]
match_df
```

```python
delivery_df = match_df.merge(delivery, on='match_id')
delivery_df=delivery_df[delivery_df['inning']==2]
delivery_df
```

```python
delivery_df['current_score'] = delivery_df.groupby('match_id').agg({'total_runs_y': 'cumsum'})
# delivery_df['total_runs_y'].dtypes
delivery_df
```

```python
delivery_df['run_left'] = delivery_df['total_runs_x'] - delivery_df['total_runs_y']
```

```python
delivery_df
```

```python
delivery_df['balls_left'] = 126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna('0')
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
wickets= delivery_df.groupby('match_id').agg({'player_dismissed': 'cumsum'})
# delivery_df['wickets'] = 10 - wickets
delivery_df.tail()
print(wickets)
```
